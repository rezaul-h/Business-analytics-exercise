```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#location of the IBM data 
path = 'data/'
```

```python
#location of all the attribute files
attribute1 = path + 'USPTO_2012_1_proj_attributes_NEW.csv' 
attribute2 = path + 'USPTO_2012_2_proj_attributes_NEW.csv'
attribute3 = path + 'USPTO_2012_3_proj_attributes_NEW.csv'
attribute4 = path + 'USPTO_2012_4_proj_attributes_NEW.csv'
attribute5 = path + 'USPTO_2012_5_proj_attributes_NEW.csv'
```

```python
#location of all the edgelist files
edgelist1 = path + 'USPTO_2012_1_proj_edgelist_NEW.csv' 
edgelist2 = path + 'USPTO_2012_2_proj_edgelist_NEW.csv'
edgelist3 = path + 'USPTO_2012_3_proj_edgelist_NEW.csv'
edgelist4 = path + 'USPTO_2012_4_proj_edgelist_NEW.csv'
edgelist5 = path + 'USPTO_2012_5_proj_edgelist_NEW.csv'
```

```python
#read all the csv files of attributes 
attribute_df1 = pd.read_csv(attribute1)
attribute_df2 = pd.read_csv(attribute2)
attribute_df3 = pd.read_csv(attribute3)
attribute_df4 = pd.read_csv(attribute4)
attribute_df5 = pd.read_csv(attribute5)
```

```python
final_attribute = pd.concat([attribute_df1, attribute_df2, attribute_df3, attribute_df4, attribute_df5], ignore_index=True)
final_attribute = final_attribute.drop(columns=['Unnamed: 10'])
print('total shape of attributes {}'.format(final_attribute.shape))
```

```python
#read all the csv files of attributes 
colnames = ['ID', 'Multinational2']
edgelist_df1 = pd.read_csv(edgelist1,names=colnames)
edgelist_df2 = pd.read_csv(edgelist2,names=colnames)
edgelist_df3 = pd.read_csv(edgelist3,names=colnames)
edgelist_df4 = pd.read_csv(edgelist4,names=colnames)
edgelist_df5 = pd.read_csv(edgelist5,names=colnames)
```

```python
final_edgelist = pd.concat([edgelist_df1,edgelist_df2,edgelist_df3,edgelist_df4,edgelist_df5] ,ignore_index=True)

print('total shape of attributes {}'.format(final_attribute.shape))
```

```python
df = pd.merge(final_attribute, final_edgelist, on='ID')
```

```python
df.tail()
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = ((df.isnull().sum()/df.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('data destribution of categorical columns')

# Patent_industry
sns.barplot(ax=axes[0], x=df.Patent_industry.value_counts().index, y=df.Patent_industry.value_counts().values)
axes[0].set_title('Patent_industry')


# University
sns.barplot(ax=axes[1], x=df.University.value_counts().index, y=df.University.value_counts().values)
axes[1].set_title('University')

# Country_code
sns.barplot(ax=axes[2], x=df.Country_code.value_counts().index, y=df.Country_code.value_counts().values)
axes[2].set_title('Country_code');

```

```python
# Patent_industry(1-5), University(0-1), Country_code(1-11) categorical column

print(df.Patent_industry.value_counts())
print(df.University.value_counts())
print(df.Country_code.value_counts())
```

```python
fig, axes = plt.subplots(3, 2, figsize=(18, 10))
fig.suptitle('Numerical columns data destribution')

# Patent_count
sns.distplot( df['Patent_count'],ax = axes[0, 0])

# Turnover_lay
sns.distplot(df['Turnover_lay'],ax = axes[0, 1])

# Turnover_2012
sns.distplot(df['Turnover_2012'],ax = axes[1, 0])

# Total_assets_2012
sns.distplot(df['Total_assets_2012'],ax = axes[1, 1])

# R&D_2012
sns.distplot(df['R&D_2012'],ax = axes[2, 0]);
```

```python
#skewness and kurtosis
print("Skewness: %f" % df['Turnover_2012'].skew())
print("Kurtosis: %f" % df['Turnover_2012'].kurt())
```
